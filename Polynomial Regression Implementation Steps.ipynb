```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import warnings

from sklearn.preprocessing import PolynomialFeatures

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('position_salaries.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,1:2].values
y=df.iloc[:,2].values
```

```python
2
```

```python
model_lr=LinearRegression()
model_lr.fit(x,y)
```

```python
model_pol_2=PolynomialFeatures( degree = 2 )

x_poly_2 = model_pol_2.fit_transform(x)

model_lr_2=LinearRegression()
model_lr_2.fit(x_poly_2,y)

model_pol_3=PolynomialFeatures( degree = 3 )

x_poly_3 = model_pol_3.fit_transform(x)

model_lr_3=LinearRegression()
model_lr_3.fit(x_poly_3,y)
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,model_lr.predict(x),color='g')
plt.title('Simple LR')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.style.use('fivethirtyeight')

plt.scatter(x,y,color='r')

plt.plot(x, model_lr_2.predict(model_pol_2.fit_transform(x)), color='g')
plt.plot(x, model_lr_3.predict(model_pol_3.fit_transform(x)), color='pink')

plt.title('Polynomial Linear Regression Degree 2,3')

plt.legend='upper left'

plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.show()
```

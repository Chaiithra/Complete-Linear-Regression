```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model, metrics 
```

```python
data_url = 'http://lib.stat.cmu.edu/datasets/boston'
raw_df=pd.read_csv(data_url, sep='\s+', skiprows=22,  header=None)
```

Pre-Processing Data

```python
x=np.hstack([raw_df.values[::2,:], raw_df.values[1::2,:2]])
y=raw_df.values[1::2,2]
```

Splitting Data into Training and Testing Sets

```python
x_train,x_test,y_train,y_test = train_test_split( x, y, test_size = 0.4, random_state = 1 )
```

Creating and Training LR Model

```python
model = linear_model.LinearRegression()
model.fit(x_train, y_train)
```

Evaluation Model Performance

```python
print('Coefficients: ', model.coef_)

print('Variance Score: {}'.format(model.score(x_test,y_test)))
```

Here if Variance score = 1 Then it's a perfect prediction and maybe a case of overfitting too 

Plotting Residual errors

```python
plt.style.use('fivethirtyeight')

# Residual Errors in Training Data 
plt.scatter(model.predict(x_train), model.predict(x_train)-y_train, color='g', s=10, label='Training Data')

# Residual Errors in Test Data
plt.scatter(model.predict(x_test), model.predict(x_test)-y_test, color='b', s=10, label='Test Data')

# Plotting Line fro Zero Residual Error
plt.axhline(y=0,xmin=0,xmax=50, linewidth=2)

plt.legend(loc='upper right')
plt.title('Residual Errors')

plt.show()
```
